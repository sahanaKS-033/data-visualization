Calculate average milk production by cow, buffalo and goat in each year for selected state.

```python
ks=input("Enter state ")
df['mean 10-11']=df[['Cow Milk-2010-11', 'Boffalo Milk-2010-11', 'Goat Milk-2010-11']].mean(1)
df['mean 11-12']=df[['Cow Milk-2011-12', 'Boffalo Milk-2011-12', 'Goat Milk-2011-12']].mean(1)
df['mean 13-14']=df[['Cow Milk-2013-14', 'Boffalo Milk-2013-14', 'Goat Milk-2013-14']].mean(1)
df['mean 14-15']=df[['Cow Milk-2014-15', 'Boffalo Milk-2014-15', 'Goat Milk-2014-15']].mean(1)
df['mean 15-16']=df[['Cow Milk-2015-16', 'Boffalo Milk-2015-16', 'Goat Milk-2015-16']].mean(1)
df[df['State/ UT Name']==ks][['State/ UT Name','mean 10-11','mean 11-12','mean 13-14','mean 14-15','mean 15-16']]
```

List 5 states (if present) whose total milk production is increased for last three years.

```python
c = df.groupby('State/ UT Name')['Cow Milk-2011-12','Boffalo Milk-2011-12','Goat Milk-2011-12'].transform(sum,axis=1,ascending=True)
d=df.groupby('State/ UT Name')['Cow Milk-2014-15','Boffalo Milk-2014-15','Goat Milk-2014-15'].transform(sum,axis=1,ascending=True)
f=df.groupby('State/ UT Name')['Cow Milk-2013-14','Boffalo Milk-2013-14','Goat Milk-2013-14'].transform(sum,axis=1,ascending=True)
print(f.head())
print(d.head())
print(c.head())
```
